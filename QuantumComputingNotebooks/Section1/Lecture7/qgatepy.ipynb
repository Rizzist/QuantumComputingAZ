```python
import numpy as np
```

```python
a = 1 + 1j
b = 1 - 1j
```

```python
x = np.array([a, b])
```

```python
def normalize(vector):
    def coeffLength(coeff):
        return np.real(coeff*np.conj(coeff))
    def totalLength():
        sum = 0
        for i in vector:
            sum += coeffLength(i)
        return sum**(0.5)
    totalSum = np.array(totalLength())
    return vector/totalSum
```

```python
normalize(x)
```

```python
def innerproduct(A, B):
    return B*np.matrix(A).getH()
```

```python
print(innerproduct(x, x))
```

```python
def outerproduct(A, B):
    return np.transpose(np.matrix(A).getH()*B)
```

```python
outerproduct(x, x)
```

```python
#unitary, kroneckor product
```

```python
M = np.array(
    [
        [1, 1],
        [1, -1],
    ])
```

```python
def isUnitary(M):
    return np.allclose(np.eye(M.shape[0]), outerproduct(M, M))
```

```python
isUnitary(M)
```

```python
def tensordot(A, B):
    return np.kron(A, B)
```

```python
tensordot(M, M)
```
